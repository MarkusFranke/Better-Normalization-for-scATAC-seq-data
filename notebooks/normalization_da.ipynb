# Simulate differential accessibility

## Intro
### High peak variance as an indicator of differential accessibility
- One of the most important goals of count normalization is the stabilization of variance.
- Ideally peaks that show high variance would also be diffentially accessible (DA).
- To test for that we will simulate differentially accessible peaks using simATAC.

```python
perf_dict = {result.key: result.performance for result in results}
df = pd.DataFrame.from_dict(perf_dict, orient="index")
```

```python
df.loc[:, 'method'] = df.loc[:, 'method'].str.replace('_normalization','')
```

```python
del df['feature_group']
del df['run']
del df['fdr']
```

```python
group=['cell_group', 'method', 'scenario']
stats_df = df.groupby(group).agg(['mean', 'count', 'std'])
for column in set(stats_df.columns.get_level_values(0)):
    stat = stats_df.loc[:, column]
    ci95 = []

    for i in stat.index:
        m, c, s = stat.loc[i]
        ci95.append(1.96*s/math.sqrt(c))
    stats_df[(column, 'ci95')] = ci95
```

```python
stats_df.reset_index(inplace=True)
```

## Scenario 1: Same library size
That includes cases S0, S2 and S4
We will only look at C1

```python
scenarios = {'S2': 'low DA', 'S4': 'high DA'}
cell_groups = ['C_1']

tmp_df = df.copy()[df.cell_group.isin(cell_groups) & df.scenario.isin(scenarios.keys())]
tmp_stats = stats_df.copy()[stats_df.cell_group.isin(cell_groups) & stats_df.scenario.isin(scenarios.keys())]

tmp_df.loc[:, 'scenario'] = [scenarios[x] for x in tmp_df.scenario]
tmp_stats.loc[:, 'scenario'] = [scenarios[x] for x in tmp_stats.scenario]

del tmp_df['cell_group']
del tmp_stats['cell_group']
tmp_df = tmp_df.sort_values(['scenario', 'method'])
```

```python
fig, ax = plt.subplots()

ax.grid()
sns.barplot(tmp_df, hue='method', y='f1', ax=ax, x='scenario').set(title='Performance')
ax.set_xlabel(None)
```

```python
case1_df = tmp_stats.pivot(index='method', columns='scenario', values=('precision', 'mean'))
figsize=(3,4)
fig, ax = plt.subplots(1,  figsize=figsize) 
sns.heatmap(case1_df, annot=True, ax=ax).set(title='Precision')
ax.set_xlabel(None)
ax.set_ylabel(None)
```

### Verdict
- Low DA: 
    - fail to capture most DA peaks
    - non-binarized normalizations perform slightly better
    - library size normalization performs the best
- High DA:
    - all normalization methods perform better than raw data
    - binarized scran performs best but not significantly better
    - sc-transform moderate performance

## Scenario 2: Different library size
That includes cases S3 and S5

```python
scenarios = {'S3': 'low DA', 'S5': 'high DA'}
cell_groups = {'C_1': 'lower library size', 'C_2': 'higher library size'}
methods = {'raw': 'unnormalized', 'binarized_log1pPF': 'binary_library_log', 'raw_log1pPF': 'library_log', 'raw_PFlog1pPF': 'library_log_library', 'scTransform': 'scTransform', 'raw_scran': 'scran'}

tmp_df = df.copy()[df.cell_group.isin(cell_groups) & df.scenario.isin(scenarios.keys()) & df.method.isin(methods.keys())]
tmp_stats = stats_df.copy()[stats_df.cell_group.isin(cell_groups) & stats_df.scenario.isin(scenarios.keys()) & stats_df.method.isin(methods.keys())]

tmp_df.loc[:, 'scenario'] = [scenarios[x] for x in tmp_df.scenario]
tmp_stats.loc[:, 'scenario'] = [scenarios[x] for x in tmp_stats.scenario]
tmp_df.loc[:, 'cell_group'] = [cell_groups[x] for x in tmp_df.cell_group]
tmp_stats.loc[:, 'cell_group'] = [cell_groups[x] for x in tmp_stats.cell_group]
tmp_df.loc[:, 'method'] = [methods[x] for x in tmp_df.method]
tmp_stats.loc[:, 'method'] = [methods[x] for x in tmp_stats.method]


tmp_df = tmp_df.sort_values(['cell_group', 'scenario', 'method'])
tmp_stats = tmp_stats.sort_values(['cell_group', 'scenario', 'method'])
```

```python
# figsize=(7,12)
# y = len(cell_groups)
# fig, axes = plt.subplots(y,  figsize=figsize)

i = -1
for scenario in scenarios.values():
    i += 1
    figsize=(8, 5)
    fig, ax = plt.subplots(figsize=figsize)
    ax.grid()
    sns.barplot(tmp_df[tmp_df.scenario == scenario], hue='method', y='f1', ax=ax, x='cell_group').set(title='F1 score')
    ax.set_ylabel(scenario)
    ax.set_xlabel(None)
```

```python
# figsize=(3,7)
for met, met_name in {'f1': 'F1 score', 'precision': 'Precision', 'recall': 'Recall'}.items():
    y = len(cell_groups)
    fig, axes = plt.subplots(1, y, sharey=True)
    fig.suptitle(met_name)
    for i, cg in enumerate(cell_groups.values()):
        figsize=(3,4)
        ax = axes[i]

        mean_df = tmp_stats.set_index('cell_group').loc[cg,:].pivot(index='method', columns=['scenario'], values=(met, 'mean'))
        ci_df = tmp_stats.set_index('cell_group').loc[cg,:].pivot(index='method', columns=['scenario'], values=(met, 'ci95'))
        # fig, ax = plt.subplots()
        annot = pd.concat([mean_df, ci_df], axis=1, join='inner')
        annot = annot.apply(lambda x: pd.Series({annot.columns[0]: f'{x[0]:.2f}  {x[2]:.2f}', 
                                                   annot.columns[1]: f'{x[2]:.2f}  {x[3]:.2f}'}), axis=1)
        sns.heatmap(mean_df, fmt='', annot=annot, ax=ax).set(title=cg)
        ax.set_xlabel(None)
        ax.set_ylabel(None)
```

```python
i = -1
for cg in cell_groups.values():
    formatted = []
    i += 1
    for 
    mean_df = tmp_stats.set_index('cell_group').loc[cg,:].pivot(index='method', columns=['scenario'], values=('precision',  'mean'))
    ci_df = tmp_stats.set_index('cell_group').loc[cg,:].pivot(index='method', columns=['scenario'], values=('precision',  'ci95'))
    for col in mean_df.columns:
        formatted.append(f'{mean_df.loc["binary_library_log", col]:.2f}  {ci_df.loc["binary_library_log", col]:.2f}')
    table = pd.DataFrame({'': mean_df.columns, 'Mean  SE': formatted})
    fig, ax = plt.subplots(figsize=(3, 3))
    ax.axis('off')
    ax.axis('tight')
    ax.table(cellText=table.values, colLabels=table.columns, loc='center')
    plt.show()
```

- higher lb:
